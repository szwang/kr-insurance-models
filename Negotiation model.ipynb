```python
import numpy as np
from scipy import optimize
import math
import matplotlib.pyplot as plt
```

```python
# utility function
def utility(x):
    return .5*np.sqrt(x)

# ransom, R
def ransom(N, N_bar):
    return 5000 + (60-N) + .5*(50-N_bar)

# insurance premium, P
def premium(p, R):
    return p*R

# probability of death, q
def p_death(N):
    return N / 100+N
```

```python
# utility function properties

x = np.arange(0, 500, 0.01)
u_y = utility(x)
plt.plot(x, u_y)
plt.title('U(x)')
plt.xlabel('x')
plt.ylabel('utility')
plt.show()
```

```python
# ransom function properties

r_y = ransom(x, 30)
plt.plot(x, r_y)
plt.title('R(N, 60)')
plt.ylabel('ransom')
plt.xlabel('negotiation')
plt.show()
```

```python
# death function properties

d_y = p_death(x)
plt.plot(x, d_y)
plt.title('q(N)')
plt.xlabel('negotiation')
plt.ylabel('probability of death')
plt.show()
```

```python
# expected utility function with p, W, D, K, N, N_bar as exogenous variables
def expected_utility(p, W, D, K, N, N_bar):
    P = premium(p, ransom(N, N_bar))
    print('premium', P)
    print('W-P', W-P)
    return (1-p)*utility(W-P) + \
            p*p_death(N)*utility(W-P-D) + \
            p*(1-p_death(N))*utility(W-P-K)
```

```python
# setting exogenous variables
p = .15
W = 100
D = 100
K = 100
N_bar = 100

# function of N
def f(N):
    return expected_utility(p, W, D, K, N, N_bar)
```

```python
f(50)
```

```python
# plot function
xvals = np.arange(0, 1200, 0.01)
yvals = f(xvals)
plt.plot(xvals, yvals)
plt.xlabel('negotiation')
plt.ylabel('utility')
plt.show()
```

```python
result = optimize.minimize_scalar(f)
result
```
